a. Create the function named *astro_now*() to obtain data of the astraonauts in the space using the method [Number of People in Space](http://open-notify.org/Open-Notify-API/People-In-Space) using the following python libraries:

* [json](https://docs.python.org/3/library/json.html)

```python
# Complete task a here
def astro_now():
    """Transforms an API url to a dictionary.
    
    In this function we take the url of the API of 
    austronauts in spance and transform it to a 
    dictionary.
    
    Args: 
        None.
    
    Returns: 
        a dictionary with the content of the API.
    """
    endpoint_url = 'http://api.open-notify.org/astros.json' #assign the url to a varible to better use
    res = requests.get(endpoint_url) #We need to know if it would be txt, json, etc
    results = json.loads(res.content) #res is an object, but with content we can see it
    return results
```

c. Create the function named *iss_currentlocation*() to obtain the position of the International Space Station using the method [Current Location of the International Space Station](http://open-notify.org/Open-Notify-API/ISS-Location-Now/).

d. Use your *iss_currentlocation()* function to obtain the position of the International Space Station each 10 seconds during 5 minutes. Show the results in a table with the following columns:
* timestamp (Parsed Correctly)
* latitude
* longitude

The [Time](https://docs.python.org/3/library/time.html) Python library could be useful.

```python
def iss_loc10():
    """Create a data frame with the lat,loc and
    timestamp every 10 sec for 5 min.
    
    In this function we take use the iss_currentlocation()
    and iss_timestamp() functions, add them in a list, so
    that we can create the data frame with the latitude,
    longitude and timestamp, but the main it's that is done 
    every 10 seconds for 5 minutes.
    
    Args: 
        None.
    
    Returns: 
        A data frame with the lat, long, and timestamp.
    """
    data = [] #empty list
    minutes = 0 #inizialitiar
    while minutes < 300: #for the 5 min
        location = iss_currentlocation() #calling the function and assign it to a variable
        data.append(location) #Adding to the list the locaation that already includes the timestamp
        df = pd.DataFrame(data=data, columns=['latitude','longitude','timestamp']) #creating the data frame
        time.sleep(10) #to make it run every 10 sec
        minutes += 10 #to make it to the 5 min
    df['timestamp'] = pd.to_datetime(df['timestamp'],unit='s') #changing to seconds the timestamp
    return df    
```

e. The International Space Station is an orbital outpost circling high above out heads. Sometimes itâ€™s overhead, but when? It depends on your location. Given a location on Earth (latitude, longitude, and altitude) the [International Space Station Pass Times](http://open-notify.org/Open-Notify-API/ISS-Location-Now/) API compute the next n number of times that the ISS will be overhead.

The function must return the obtained json file as a python dictionary.

f. Write the *url request* to obtaing the same result as in task e.

```python
def iss_passtimes (latitude, longitude, altitude = None, number = None):
    """Transforms an API url to a dictionary.
    
    In this function we take the url of the API of 
    when and the duration of The International Space 
    Station would be above our heads, and transform 
    it to a dictionary.
    
    Args: 
        latitude: of the location desired.
        longitude: of the location desired.
        altitude: None
        number: none
    
    Returns: 
        A dictionary with the information.
    """
    endpoint_url = 'http://api.open-notify.org/iss-pass.json' #assign the url to a varible to better use
    parameters = {
        'lat' : latitude, 
        'lon' : longitude,
        'alt' : altitude, 
        'n' : number
    }
    
    res = requests.get(endpoint_url, params = parameters) 
    results = json.loads(res.content)
    return results
```
